```python
#upload the dataset we want to use
final = pd.read_csv('emotions_with_neutrals_220228.csv')
```

Tokenize and Lemmatize the words,
get rid of stop words excluding the negative stop words since negation is important for emotion detection.

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
import re
from nltk import corpus
from nltk import WhitespaceTokenizer
from nltk import WordNetLemmatizer
#nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
#engstop_path =  '/root/nltk_data/corpora/stopwords/english'
engstop = stopwords.words('english')

negation = ['no','nor','not','don',"don't",'should',"should've",'ain',\
            'aren',"aren't",'couldn',"couldn't",'didn',"didn't",'doesn',\
            "doesn't",'hadn',"hadn't",'hasn',"hasn't",'haven',"haven't",\
            'isn',"isn't","mightn't",'mustn',"mustn't",'needn',"needn't",\
            'shan',"shan't",'shouldn',"shouldn't",'wasn',"wasn't",'weren',\
            "weren't",'won',"won't",'wouldn',"wouldn't"]
stop = [i for i in engstop if i not in negation]

def clean_ats(content):
    content = content.lower()
    content = re.sub("'", "", content)
    content = re.sub("@[A-Za-z0-9_]+","", content)
    content = re.sub("&[A-Za-z0-9_]+","", content)
    content = re.sub('[()!?]', ' ', content)
    content = re.sub('\[.*?\]',' ', content)
    content = re.sub('\s+',' ', content)
    content = re.sub('[\r\n]+', ' ', content)
    content = re.sub('&amp', 'and', content)
    content = re.sub('&lt', '<', content)
    content = re.sub('&gt', '>', content)
    content = re.sub("[^a-zA-Z]"," ", content)
    return content
final['cleaned'] = final['text'].apply(clean_ats)

#final['cleaned_stopwords'] = final['lemmatized'].apply(lambda x: [word for word in x if word not in (stop)])
final
```

```python
lemmatizer = WordNetLemmatizer()
w_tokenizer = WhitespaceTokenizer()
def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
final['lemmatized'] = final['cleaned'].apply(lemmatize_text)


def tokenizer(text):
  return w_tokenizer.tokenize(text)

  
#final['tokenized'] = final['cleaned'].apply(tokenizer)
final
```

Check lemmatization works as expected

```python
final['cleaned_stopwords'] = final['lemmatized'].apply(lambda x: [word for word in x if word not in (stop)])
print("abused :", lemmatizer.lemmatize("abused"))
print("acquired :", lemmatizer.lemmatize("acquired"))
```

```python
use = final[['cleaned_stopwords', 'emotions']]
use['emotions'].unique()
use
```

Explode cleaned words and find count by emotions and overall

```python
use = use.explode('cleaned_stopwords')
t = use.copy()
t['counter'] = 1
j = t.groupby(['cleaned_stopwords', 'emotions']).agg('count').reset_index()
k = j.groupby(['cleaned_stopwords']).agg('sum').reset_index()
keywordcounts = pd.merge(j, k, how='left', left_on='cleaned_stopwords', right_on='cleaned_stopwords')
keywordcounts
```

Filter and save

```python
keywordcounts['percent_of_group'] = keywordcounts['counter_x']/keywordcounts['counter_y']
lexicon = keywordcounts.query('counter_y >= 30 & percent_of_group >= 0.65')

#keywordcounts[[keywordcounts['counter_y']>20  & keywordcounts['percent_of_group']>0.5]]
lexicon
lexicon.to_csv('lexicon_220306.csv')
files.download('lexicon_220306.csv')
```
